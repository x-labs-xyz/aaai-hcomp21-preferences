# Geolocation

```python
# import map data into states dataframe
fp = 'ip-map-datasets/usa-states-census-2014.shp'
states = gpd.read_file(fp)
states = states.to_crs("EPSG:3395")
```

```python
# map the number of users from each state to the map data
users['state_count'] = users.groupby('user_state')['user_state'].transform('count')
state_dict = dict(zip(users.user_state, users.state_count))
states['state_count'] = states.NAME.map(state_dict)
```

```python
# plot choropleth map
variable = 'state_count'
vmin, vmax = states['state_count'].min(), states['state_count'].max()
fig, ax = plt.subplots(figsize=(15, 15))
states.plot(column=variable, cmap='BuPu',linewidth=0.8,edgecolor='0.8', ax=ax)

ax.axis('off')
sm = plt.cm.ScalarMappable(cmap='BuPu', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)

fig.colorbar(sm, cax=cax)
```
